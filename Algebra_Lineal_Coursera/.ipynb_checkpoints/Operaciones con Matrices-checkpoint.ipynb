## IV Multiplicacion de Matrices

### 1. Inner (or Dot) Product --> producto interno o puntual


#### Tenemos dos vectores columnas: v y u. Entonces el producto interno de uv, se denota por v transpuesta x u.


```python
# Ejemplo:

v = np.array([[1],
              [0],
              [-1],
              [5]])
u = np.array([[2],
              [1],
              [0],
              [-1]])
print(v)
print()
print(u)
```

```python
v_trans = v.T
v_trans
```

```python
result = v_trans @ u
print("El producto interno de los vectores es:",result, sep="\n")
```

```python
# Otra forma:
# Usamos ravel si deseamos aplanar un vector matriz, es decir, llevarlo a una forma de lista, sin filas ni columnas.
np.inner(v.ravel(),u.ravel())
```

```python
# Otra forma:
np.dot(v.ravel(), u.ravel())  #  -3
```

### 2. producto de matrices
#### El producto de matrices se realiza multiplicando la primera fila de la matriz A por la primera columna de la matriz B, asi para todos los valores.

```python
A = np.array([[1,1],
              [1,0]])
print(A)
print()

B = np.array([[0,1],
              [1,2]])
print(B)
print()
C = A @ B

print(C)
```

```python
# Recordemos que en la multiplicacion AB != BA, el producto no es commutativo
# Ademas, el numero de columnas de la primera matriz, debe ser igual al numero de filas de la segunda matriz
A = np.array([[-1,0],
              [0,5]])
print(A)
print()
B = np.array([[1,1],
              [3,-5]])
print(B)
print()

print("El producto A@B es :",A@B, sep="\n")
print()
print("El producto B@A es :",B@A, sep="\n")
```

```python
# Generamos una matriz identidad de 2x2

I = np.eye(2)
print(C)
```

```python
# Multiplicamos por A
D = I @ A
E = A @ I
print("Multiplicacion de I@A",D,sep="\n")
print()
print("Multiplicacion de A@I",E,sep="\n")
```

#### Ejemplo practico

#### Solo tiene tiempo para visitar una tienda:
#### Opciones: tienda A,B o C.
#### Necesitas: Brocoli, Tomates y zanahorias.
#### Objetivo: gastar lo minimo en las compras.

#### Para la cena necesitamos 2.5 kg de brocoli, 4 kg de tomates y 2 kg de zanahorias.
#### Pregunta:  Que tienda debo visitar para gastar el menor dinero posible?
#### Los precios son los siguientes:


```python
# Generamos la matriz de precios
P = np.array([[1.49,3.99,2.49], # Precios de la tienda A
              [1.69,2.49,1.69], # Precios de la tienda B
              [1.04,3.52,1.28]]) # Precios de la tienda C
# Generamos la matriz de incognitas que ya son dadas.
R = np.array([[2.5],
              [4.],
              [2.]])


print(P)
print()
print(R)
```

```python
# Si deseamos saber la mejor opcion de compra debemos mutiplicar P.R

res = np.dot(P,R)
print(res)
```

```python
# Entonces el mejor precio seria de la tienda B
print(f"El mejor precio es de la tienda B, con:  {res[1]}")
```

##### Problema resuelto 2 


```python
A_2 = np.array([[4,1,-2],
                [3,1,3],
                [6,2,1]])
B_2 = np.array([1,10,10])

C_2 = np.array(['x','y','z'])
```
